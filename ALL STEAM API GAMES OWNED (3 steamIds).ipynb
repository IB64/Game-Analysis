```python
import requests
import pandas as pd

def extract_data(api_key, steam_id):
    # API endpoint URL
    url = f"https://api.steampowered.com/IPlayerService/GetOwnedGames/v0001/?key={api_key}&steamid={steam_id}&format=json&include_appinfo=true"

    # Send GET request to the API endpoint
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the JSON response
        data = response.json()['response']['games']
        # Use pd.json_normalize() to flatten nested JSON data and create a DataFrame
        df = pd.json_normalize(data)
        
        # Convert playtime_forever from minutes to hours
        df['playtime_forever'] /= 60  # Divide by 60 to convert minutes to hours
        
        # Selecting only the columns of interest
        df = df[['appid', 'name', 'playtime_forever']]
        
        pd.set_option('display.precision', 2)
        
        # Display the DataFrame
        print(df)
    else:
        # Print an error message if the request was unsuccessful
        print("Error:", response.status_code) 

# Example of usage
if __name__ == "__main__":
    # Example configuration (you can read this from a file)
    users = [
        {"api_key": "638B25CA545D1DCFC5E2188A15A841F5", "steam_id": "76561198235391392"},
        {"api_key": "638B25CA545D1DCFC5E2188A15A841F5", "steam_id": "76561199204427262"},
        {"api_key": "638B25CA545D1DCFC5E2188A15A841F5", "steam_id": "76561198155762381"}
        # Add more users as needed
    ]
    
    for user in users:
        extract_data(user['api_key'], user['steam_id'])
```
